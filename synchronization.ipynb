```python
df = pd.read_csv("only_radar.csv")
radar_df = pd.DataFrame({'radar_timestamp': df['radar_timestamp']})
radar_df = radar_df.sort_values(['radar_timestamp'], ascending=[True])
radar_df = radar_df.reset_index(drop=True)
radar_list = radar_df['radar_timestamp'].tolist()
radar_df.to_csv("temp.csv", index=False)
radar_list
```

```python
df = pd.read_csv("only_camera.csv")
camera_df = pd.DataFrame({'camera_timestamp': df['camera_timestamp']})
camera_df = camera_df.sort_values(['camera_timestamp'], ascending=[True])
camera_df = camera_df.reset_index(drop=True)
camera_list = camera_df['camera_timestamp'].tolist()
camera_df.to_csv("temp.csv", index=False)
camera_list
```

```python
df = pd.DataFrame(columns=['camera_timestamp', 'radar_timestamp'])
df
```

```python
df['radar_timestamp'] = radar_list
df
```

```python
df['camera_timestamp'] = camera_list + [float('nan')] * (len(radar_list) - len(camera_list))
df
```

```python
df.to_csv("temp.csv", index=False)
```

```python
import numpy as np

df['camera_timestamp'] = pd.to_datetime(df['camera_timestamp'], format='%H:%M:%S.%f')
df['radar_timestamp'] = pd.to_datetime(df['radar_timestamp'], format='%H:%M:%S.%f')

df['closest_radar_timestamp'] = np.nan

for i, camera_ts in enumerate(df['camera_timestamp']):
    if pd.isnull(camera_ts):
        break
    closest_radar_idx = df['radar_timestamp'].sub(camera_ts).abs().idxmin()
    closest_radar_ts = df.loc[closest_radar_idx, 'radar_timestamp'].strftime('%H:%M:%S.%f')
    df.loc[i, 'closest_radar_timestamp'] = closest_radar_ts
```

```python
df.to_csv("Radar_Camera_Synch.csv", index=False)
```
