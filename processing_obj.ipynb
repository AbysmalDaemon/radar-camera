Transpose the data

```python
import csv

# Specify the input file path
input_file = 'trial.csv'  # Replace with the actual input file path

# Specify the output file path
output_file = 'trial1.csv'  # Replace with the desired output file name

# Read the CSV file
with open(input_file, 'r') as csv_file:
    reader = csv.reader(csv_file)
    data = list(reader)

# Transpose the data
transposed_data = list(map(list, zip(*data)))

# Write the transposed data to a new CSV file
with open(output_file, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerows(transposed_data)

print(f"Transposed data saved to: {output_file}")
```

Remove Brackets

```python
import os
import csv

# Create the 'radar1' folder if it doesn't exist
if not os.path.exists('radar1'):
    os.makedirs('radar1')

# Get a list of all CSV files in the 'radar_new' folder
csv_files = [file for file in os.listdir('trial1') if file.endswith('.csv')]

# Iterate over each CSV file
for file in csv_files:
    file_path = os.path.join('trial1', file)
    output_path = os.path.join('radar1', file)
    
    # Open the input CSV file
    with open(file_path, 'r') as input_file:
        reader = csv.reader(input_file)
        rows = list(reader)
        
        # Modify the second column of each row
        for row in rows:
            if len(row) >= 2:
                row[1] = row[1].lstrip('(').rstrip(')')
        
    # Save the modified data to a new CSV file in the 'radar1' folder
    with open(output_path, 'w', newline='') as output_file:
        writer = csv.writer(output_file)
        writer.writerows(rows)
```

Add Labels Row

```python
import os
import csv

# Specify the folder path where the CSV files are located
input_folder_path = "radar1"

# Specify the folder path to save the output CSVs
output_folder_path = "radar2"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)

# Get a list of all files in the input folder
file_list = os.listdir(input_folder_path)

# Iterate over each file in the folder
for file_name in file_list:
    if file_name.endswith(".csv"):  # Process only CSV files
        input_file_path = os.path.join(input_folder_path, file_name)
        output_file_path = os.path.join(output_folder_path, file_name)

        # Create a new row with labels and values
        new_row = ['Parameters', 'Values']

        # Read the existing CSV data
        existing_rows = []
        with open(input_file_path, 'r') as file:
            reader = csv.reader(file)
            existing_rows = list(reader)

        # Append the new row on top of the existing rows
        updated_rows = [new_row] + existing_rows

        # Write the updated rows to the output CSV file
        with open(output_file_path, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(updated_rows)
```

```python
import os
import pandas as pd

# Specify the folder path where the CSV files are located
input_folder_path = "radar3"

# Specify the folder path to save the output CSVs
output_folder_path = "radar4"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)

# Get a list of all files in the input folder
file_list = os.listdir(input_folder_path)

# Iterate over each file in the folder
for file_name in file_list:
    if file_name.endswith(".csv"):  # Process only CSV files
        input_file_path = os.path.join(input_folder_path, file_name)
        output_file_path = os.path.join(output_folder_path, file_name)
        
        df = pd.read_csv(input_file_path, on_bad_lines='skip')
        second_column = df.iloc[:, 1]  # Assuming the second column index is 1 (0-indexed)
        values = second_column.tolist()
        print(len(values))

        with open(output_file_path, 'w') as file:
            for value in values:
                file.write(str(value))
                file.write('\n')
```
